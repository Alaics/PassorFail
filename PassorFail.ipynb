```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score



df = pd.read_csv("https://raw.githubusercontent.com/ShapeLab/ZooidsCompositePhysicalizations/refs/heads/master/Zooid_Vis/bin/data/student-dataset.csv")


print(df.shape)
```

```python

df.head()
```

```python
df = df[['english.grade', 'math.grade', 'sciences.grade', 'language.grade']]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python

failing_threshold = 3


df['Pass/Fail'] = np.where((df['english.grade'] > failing_threshold) &
                            (df['math.grade'] > failing_threshold) &
                            (df['sciences.grade'] > failing_threshold) &
                            (df['language.grade'] > failing_threshold), 0, 1)
```

```python
X = df.iloc[:, 0:-1]
Y = df['Pass/Fail']

print(X)
print(Y)
```

```python
df['Pass/Fail'].value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

precision = precision_score(y_test, y_pred)
print(f"Model Precision: {accuracy * 100:.2f}%")

recall = recall_score(y_test, y_pred)
print(f"Model Recall: {accuracy * 100:.2f}%")
```

```python
new_student_scores = pd.DataFrame([[ 5, 3, 3, 5]], columns=['english.grade', 'math.grade', 'sciences.grade', 'language.grade'])

prediction = model.predict(new_student_scores)

if prediction == 1:
    print("The student is predicted to Pass.")
else:
    print("The student is predicted to Fail.")
```
